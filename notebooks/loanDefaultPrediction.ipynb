```python
# Data Pre-Processing

# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# For SMOTE
from imblearn.over_sampling import SMOTE

# To save the scaler pipeline if needed
import joblib
```

```python
RND = 42 # Random seed constant
```

```python
# Sometimes kaggle files have an index column named 'Unnamed: 0' - drop if it is present
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
# Get the Datatype of each variable/column
df.dtypes
```

```python
# Defensive checks & column list
expected_cols = [
    'SeriousDlqin2yrs', 'RevolvingUtilizationOfUnsecuredLines', 'age',
    'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'MonthlyIncome',
    'NumberOfOpenCreditLinesAndLoans', 'NumberOfTimes90DaysLate',
    'NumberRealEstateLoansOrLines', 'NumberOfTime60-89DaysPastDueNotWorse',
    'NumberOfDependents'
]

missing_expected = [c for c in expected_cols if c not in df.columns]
if missing_expected:
    print("Warning: expected columns missing:", missing_expected)
else:
    print("All expected columns present.")
```

```python
# Handling the Missing Value (Imputations) with indicator to MonthlyIncome
# MonthlyIncome often has many missing values - create a missing indicator 

df['MonthlyIncome_missing'] = df['MonthlyIncome'].isnull().astype(int)

# Impute MonthlyIncome with median (robust)
monthly_median = df['MonthlyIncome'].median()
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(monthly_median)
```

```python
# NumberOfDependents: fill with mode (most common value) and convert to integer
if df['NumberOfDependents'].isnull().sum() > 0:
    mode_dep = df['NumberOfDependents'].mode().iloc[0]
    df['NumberOfDependents'] = df['NumberOfDependents'].fillna(mode_dep)
    
# Ensure integer type (some files have float)
df['NumberOfDependents'] = df['NumberOfDependents'].astype(int)
```

```python
# Check again if the Missing Values are available now
df.isnull().sum()
```

```python
# Get the Head of the dataset
df.head()
```

```python
# Outlier Handling / Capping (winsorization)

# Use percentile to handle extreme values

# 1. Age: reasonable bounds (18 to 99th percentile)
min_age = 18
max_age = df['age'].quantile(0.995)
df['age'] = df['age'].clip(lower=min_age, upper=max_age)
```

```python
# 2. MonthlyIncome: cap at 99th percentile
inc_99 = df['MonthlyIncome'].quantile(0.99)
df['MonthlyIncome'] = df['MonthlyIncome'].clip(upper=inc_99)
```

```python
# 3. DebtRatio: cap at 99th percentile (DebtRatio may have huge outliers)
dr_99 = df['DebtRatio'].quantile(0.99)
df['DebtRatio'] = df['DebtRatio'].clip(upper=dr_99)
```

```python
# 4. RevolvingUtilizationOfUnsecuredLines: cap at 99.5th percentile
rev_995 = df['RevolvingUtilizationOfUnsecuredLines'].quantile(0.995)
df['RevolvingUtilizationOfUnsecuredLines'] = df['RevolvingUtilizationOfUnsecuredLines'].clip(upper=rev_995)
```

```python
# 5. For count features (past due counts, open lines, realestate loans) cap at a reasonable integer (e.g., 10)
count_cols = [
    'NumberOfTime30-59DaysPastDueNotWorse','NumberOfTimes90DaysLate',
    'NumberOfTime60-89DaysPastDueNotWorse','NumberOfOpenCreditLinesAndLoans',
    'NumberRealEstateLoansOrLines'
]
for c in count_cols:
    if c in df.columns:
        df[c] = df[c].clip(upper=10)  # cap at 10 (adjust if you prefer)
```

```python
# Feature Engineering (helpful signals)

# Total past delinquencies (aggregate of 30-59, 60-89, 90+)
df['TotalPastDue'] = (
    df['NumberOfTime30-59DaysPastDueNotWorse']
    + df['NumberOfTime60-89DaysPastDueNotWorse']
    + df['NumberOfTimes90DaysLate']
)
```

```python
# Log Transforms for skewed continuous columns
# log1lp handles zero values safely

df['LogMonthlyIncome'] = np.log1p(df['MonthlyIncome'])
df['LogDebtRatio'] = np.log1p(df['DebtRatio'])
```

```python
# Income per open credit line (proxy for financial slack)
df['IncomePerOpenLine'] = df['MonthlyIncome'] / (df['NumberOfOpenCreditLinesAndLoans'] + 1)
```

```python
# Optional: if you want a binary flag for any past delinquency
df['HasPastDelinquency'] = (df['TotalPastDue'] > 0).astype(int)
```

```python
# Get the head of the dataset
df.head()
```

```python
# Prepare feature (X) and target (Y)
TARGET = 'SeriousDlqin2yrs'

# Drop any ID columns if present (none expected)
possible_id_cols = ['Id', 'ID', 'loan_id']
for c in possible_id_cols:
    if c in df.columns:
        df = df.drop(columns=[c])
        
# Create X and Y
X = df.drop(columns=[TARGET])
y = df[TARGET].astype(int)

print("X-Shape: ", X.shape)
print("y Distribution:\n ", y.value_counts(normalize=True))
```

```python
# Train-test split (stratified because target is imbalanced)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=RND)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train target distribution:\n", y_train.value_counts(normalize=True))
print("Test target distribution:\n", y_test.value_counts(normalize=True))
```

```python
# Scaling with numeric feature

# Decide which column to scale (numerical column)
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()

# It's ok to scale all numerical features; keep track of column order
print("Numeric columns to scale:", numeric_cols)
```

```python
# Do the Standardization
scaler = StandardScaler()

# Fit on training numeric feature
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])

# Save scaler for later (so you can scale new data the same way)
joblib.dump(scaler, "Models/scaler_gmsc.joblib")
print("Scaling complete. Scaler saved to Models/scaler_gmsc.joblib")
```

```python
# Handling the class imbalance

# Approach A: Use class weights in classifier (no data resampling needed)
# Example: in LogisticRegression use `class_weight='balanced'` or compute weights manually from y_train

from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_train)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weight_dict = dict(zip(classes, class_weights))
print("Class weights (to pass to classifiers):", class_weight_dict)
```

```python
# Approach B: Resample training set using SMOTE (only fit on training data)

sm = SMOTE(random_state=RND)
X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train)

print("After SMOTE - X_train_res shape:", X_train_res.shape)
print("After SMOTE - class distribution:\n", pd.Series(y_train_res).value_counts(normalize=True))
```

```python
# Save preprocessed datasets for modeling
# (save both scaled original and SMOTE-resampled versions)

X_train_scaled.to_csv("Data/X_train_scaled.csv", index=False)
X_test_scaled.to_csv("Data/X_test_scaled.csv", index=False)
pd.DataFrame({'y_train': y_train}).to_csv("Data/y_train.csv", index=False)
pd.DataFrame({'y_test': y_test}).to_csv("Data/y_test.csv", index=False)

# Save SMOTE-resampled arrays (they are numpy arrays; convert to DataFrame preserving column names)
X_train_res_df = pd.DataFrame(X_train_res, columns=X_train_scaled.columns)
X_train_res_df.to_csv("Data/X_train_smote.csv", index=False)
pd.DataFrame({'y_train_res': y_train_res}).to_csv("Data/y_train_smote.csv", index=False)

print("Saved preprocessed files under /Data")
```

```python
#  Quick checks & visualization (optional)
# Compare a few columns before/after scaling to verify transformations

display(X_train.head().assign(_source="original").head())
display(X_train_scaled.head().assign(_source="scaled").head())

# Check correlation of important new features with target (optional)
corr_with_target = pd.concat([X_train_scaled, y_train.reset_index(drop=True)], axis=1).corr()[TARGET].sort_values(ascending=False)
print("Correlation of features with target:\n", corr_with_target)
```

```python
# Modeling : Baseline Models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

# Use either class_weight or SMOTE-resampled data
# Here we'll demonstrate with class_weight first
models = {
    "Logistic Regression": LogisticRegression(
        class_weight=class_weight_dict, random_state=RND, max_iter=500
    ),
    "Decision Tree": DecisionTreeClassifier(
        class_weight=class_weight_dict, random_state=RND, max_depth=5
    )
}
```

```python
results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:,1]

    # Metrics
    auc = roc_auc_score(y_test, y_proba)
    print(f"{name} ROC-AUC: {auc:.3f}")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Default","Default"], yticklabels=["No Default","Default"])
    plt.title(f"{name} - Confusion Matrix")
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")

    results[name] = auc
```

```python
# Plot ROC curve comparison
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.show()

print("\nModel AUC Results:", results)
```

```python
# =========================
# MODELING: ENSEMBLE MODELS
# =========================

from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
import xgboost as xgb

# Initialize models
models_strong = {
    "Random Forest": RandomForestClassifier(
        class_weight=class_weight_dict, 
        n_estimators=200, max_depth=10, random_state=RND, n_jobs=-1
    ),
    "XGBoost": xgb.XGBClassifier(
        random_state=RND, n_estimators=300, learning_rate=0.1, 
        max_depth=6, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=class_weight_dict[0]/class_weight_dict[1], 
        eval_metric="logloss", use_label_encoder=False, n_jobs=-1
    ),
    "LightGBM": lgb.LGBMClassifier(
        random_state=RND, n_estimators=300, learning_rate=0.1, 
        max_depth=-1, subsample=0.8, colsample_bytree=0.8, class_weight=class_weight_dict, n_jobs=-1
    )
}

results_strong = {}

for name, model in models_strong.items():
    print(f"\nTraining {name}...")
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:,1]

    # Metrics
    auc = roc_auc_score(y_test, y_proba)
    print(f"{name} ROC-AUC: {auc:.3f}")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Greens", xticklabels=["No Default","Default"], yticklabels=["No Default","Default"])
    plt.title(f"{name} - Confusion Matrix")
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")

    results_strong[name] = auc

# Final ROC comparison
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison - Ensemble Models")
plt.legend()
plt.show()

print("\nStrong Model AUC Results:", results_strong)
```

```python
# Model Comparison Table
# Merge results into one DataFrame

all_results = {**results, **results_strong}
results_df = pd.DataFrame.from_dict(all_results, orient="index", columns=["ROC-AUC"]).sort_values(by="ROC-AUC", ascending=False)

print("\n==== Final Model Performance ====")
display(results_df)
```

```python
# Feature Importance

def plot_feature_importance(model, model_name, feature_names):
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1]
    plt.figure(figsize=(8, 6))
    sns.barplot(x=importances[indices], y=np.array(feature_names)[indices], palette="viridis")
    plt.title(f"{model_name} - Feature Importance")
    plt.xlabel("Importance Score")
    plt.ylabel("Features")
    plt.show()
```

```python
feature_names = X_train_scaled.columns.tolist()

# Random Forest
plot_feature_importance(models_strong["Random Forest"], "Random Forest", feature_names)

# XGBoost
plot_feature_importance(models_strong["XGBoost"], "XGBoost", feature_names)

# LightGBM
plot_feature_importance(models_strong["LightGBM"], "LightGBM", feature_names)
```
